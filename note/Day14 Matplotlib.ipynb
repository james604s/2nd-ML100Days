# 繪圖 matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')
```

### 基本繪圖

```python
x = np.linspace(-1, 1, 50)
y = 2*x + 1
plt.plot(x, y)
plt.show()
```

### Figure 基本窗口

```python

x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure()
plt.plot(x, y1)

#figure num編號 figsize長寬
plt.figure(num = 3, figsize = (8,5))
plt.plot(x, y2)
plt.plot(x, y1, color = 'blue', linewidth = 5.0, linestyle = '--')
plt.show()
```

### 座標軸調整

```python
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2


#figure num編號 figsize長寬
plt.figure(num = 3, figsize = (8,5))
plt.plot(x, y2)
plt.plot(x, y1, color = 'blue', linewidth = 5.0, linestyle = '--')

plt.xlim((-1,2)) #取值範圍
plt.ylim((-2,3))
plt.xlabel('I am x.') #xy軸調整
plt.ylabel('I am y.')

new_ticks = np.linspace(-1, 2, 5)
print(new_ticks)
plt.xticks(new_ticks) #調整x座標值
plt.yticks([-2, -1.8, -1, 1.22, 3,] ##對應文字
          ,['really bad', 'bad', 'normal', 'good', 'really good'])

#gca = 'get current axis'調整座標軸位置
ax = plt.gca()
ax.spines['right'].set_color('none') #右邊線
ax.spines['top'].set_color('none') #上邊線

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data', -1)) #x軸底部設定
ax.spines['left'].set_position(('data', 0)) #y軸底部設定


plt.show()
```

### 圖例 legend 

```python

x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure()
plt.plot(x, y1)

#figure num編號 figsize長寬
plt.figure(num = 3, figsize = (8,5))
plt.plot(x, y2 ,label = 'y2')
plt.plot(x, y1, color = 'blue', linewidth = 5.0, linestyle = '--', label = 'y1')
plt.legend(labels = ['yy2','yy1'],loc = 'best')
plt.show()
```

```python
"""
Please note, this script is for python3+.
If you are using python2+, please modify it accordingly.
Tutorial reference:
http://www.scipy-lectures.org/intro/matplotlib/matplotlib.html
Mathematical expressions:
http://matplotlib.org/users/mathtext.html#mathtext-tutorial
"""

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 50)
y = 2*x + 1

plt.figure(num=1, figsize=(8, 5),)
plt.plot(x, y,)

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

x0 = 1
y0 = 2*x0 + 1
plt.plot([x0, x0,], [0, y0,], 'k--', linewidth=2.5)
plt.scatter([x0, ], [y0, ], s=50, color='b')

# method 1:
#####################
plt.annotate(r'$2x+1=%s$' % y0, xy=(x0, y0), xycoords='data', xytext=(+30, -30),
             textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=.2"))

# method 2:
########################
plt.text(-3.7, 3, r'$This\ is\ the\ some\ text. \mu\ \sigma_i\ \alpha_t$',
         fontdict={'size': 16, 'color': 'r'})

plt.show()
```

# Subplot 多圖顯示

```python
import matplotlib.pyplot as plt


# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure()

plt.subplot(2, 2, 1)  ##給定位置描述 2行2列的第一個位置
plt.plot([0,1], [0,1])

plt.subplot(2, 2, 2)  ##給定位置描述 2行2列的第二個位置
plt.plot([0,1], [0,1])

plt.subplot(2, 2, 3)  ##給定位置描述 2行2列的第三個位置
plt.plot([0,1], [0,1])

plt.subplot(2, 2, 4)  ##給定位置描述 2行2列的第四個位置
plt.plot([0,1], [0,1])

plt.show()
```

```python
#兩列三行 第一列佔了三個位置 因此第二列為
plt.subplot(2,1,1)
plt.plot([0,1],[0,1])

plt.subplot(2,3,4)
plt.plot([0,1],[0,2])

plt.subplot(235)
plt.plot([0,1],[0,3])

plt.subplot(236)
plt.plot([0,1],[0,4])

plt.show()  # 展示
```
