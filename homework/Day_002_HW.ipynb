## select by label : loc

```python
#根據標籤選資料
print(df.loc[:,['TARGET','CNT_CHILDREN']])  #此兩欄位的全部資料
print(df.loc[3,['TARGET','CNT_CHILDREN']])  #第三列 此兩欄位的資料
```

## select by position : iloc

```python
df.iloc[2,] #第三列全部
```

```python
df.iloc[2,6]#第三列第七欄的值
```

```python
df.iloc[3:5,2:4] #第4,5列,第3,4欄的值, 
```

```python
df.iloc[[0,3,5],0:3] #第1,4,6列,第1~欄的值, 
```

```python
df.iloc[0:100:5,0:3] #0~100列間每5列取一次,第1,2欄的值
```

## mixed selection : ix

```python
df.ix[:3,['NAME_CONTRACT_TYPE','TARGET']] #0~3列,跟兩欄位的資料
```

## boolean indexing

```python
df[df.AMT_CREDIT > 1500000] #欄位條件篩選
```

## 遺失值處理

```python
df.isnull() #是否有遺失值

np.any(df.isnull()) == True 
```

```python
df.dropna(axis=1, how='any') 
#0刪除行 1刪除列, 
#how = ['any','all'] 所有數據都是遺失值再刪除
#how = 'any' 只要有遺失值就刪除
```

```python
df.fillna(value = 0) #遺失值填補為0
```
